# **Lección: Módulos**

## **Objetivo**
Al finalizar esta lección, los estudiantes deben ser comprender qué son los módulos en Python, utilizar módulos integrados y externos en sus programas, crear sus propios módulos y organizarlos en paquetes.

1. Módulo de utilidades personales:

Crea un módulo que contenga funciones para sumar, restar, multiplicar, y dividir dos números.

Importa el módulo en otro archivo y utiliza las funciones.

2. Paquete de conversión de unidades:

Crea un paquete que tenga módulos para convertir unidades de longitud, peso, y temperatura.

Usa el paquete para convertir valores en un archivo principal.

3. Módulo de administración de archivos:

Crea un módulo que contenga funciones para leer y escribir archivos de texto.

Importa el módulo en un programa principal y realiza operaciones de lectura/escritura con archivos.

4. Uso de módulos integrados:

Escribe un programa que use el módulo os para listar los archivos en un directorio y mostrar su tamaño en bytes.

5. Generador de contraseñas:

Crea un programa que utilice el módulo random para generar una contraseña aleatoria con letras, números y símbolos.

6. Calculadora Modular

Crea un módulo llamado calculadora.py con las siguientes funciones: sumar, restar, multiplicar, dividir. Luego, escribe un programa que importe este módulo y permita al usuario elegir una operación matemática e ingresar dos números.

7. Módulo de Validación

Crea un módulo llamado validacion.py que contenga funciones para validar entradas de usuario, como verificar si una cadena es un número entero, si es un correo electrónico válido, o si es una dirección web válida. Usa expresiones regulares en las validaciones. Importa este módulo en un programa que le pida al usuario diferentes entradas y las valide.

8. Módulo de Conversión de Monedas

Escribe un módulo conversiones.py que tenga funciones para convertir entre diferentes monedas (por ejemplo, de euros a dólares, de dólares a yenes). Utiliza tasas de cambio fijas. Importa este módulo en un programa principal y permite al usuario ingresar una cantidad y seleccionar las monedas de origen y destino.
   
9. Módulo de Juegos

Crea un módulo juegos.py que contenga diferentes juegos simples como "adivina el número", "piedra, papel o tijera", y "lanzar una moneda". Importa este módulo en un programa principal que permita al usuario elegir a qué juego jugar.

10. Gestión de Productos

Crea un módulo productos.py que contenga funciones para agregar productos a una lista, eliminar productos y mostrar todos los productos disponibles. Cada producto tendrá nombre, precio y cantidad. Importa el módulo en un programa principal y permite al usuario gestionar los productos.

11. Módulo de Estadísticas

Crea un módulo estadisticas.py que tenga funciones para calcular el promedio, la mediana, la moda y la desviación estándar de una lista de números. Luego, importa este módulo en un programa que le pida al usuario ingresar una lista de números y le muestre las estadísticas.

12. Generador de Contraseñas

Crea un módulo generador_contraseñas.py que tenga una función para generar contraseñas aleatorias de una longitud específica, usando letras, números y símbolos. Importa este módulo en un programa y genera una contraseña para el usuario.

13. Módulo de Manipulación de Fechas

Crea un módulo fechas.py que tenga funciones para calcular la diferencia entre dos fechas, sumar o restar días a una fecha, y verificar si una fecha es válida. Usa el módulo datetime. Importa el módulo en un programa que realice varias operaciones con fechas.

14. Módulo de Cifrado

Crea un módulo cifrado.py que contenga funciones para cifrar y descifrar mensajes usando un cifrado simple como el cifrado César. Luego, importa el módulo en un programa que permita al usuario ingresar un mensaje, cifrarlo y descifrarlo.

15. Módulo de Conversión de Unidades

Escribe un módulo conversion_unidades.py que contenga funciones para convertir entre distintas unidades (por ejemplo, de metros a pies, de litros a galones, de kilogramos a libras). Importa este módulo en un programa que permita al usuario realizar varias conversiones.

16. Administrador de Tareas

Crea un módulo tareas.py que gestione una lista de tareas con descripciones y estados (completado o pendiente). El módulo debe permitir agregar, marcar como completadas y listar las tareas. Luego, importa este módulo en un programa principal para gestionar tareas desde la consola.

17. Sistema de Registro de Ventas

Crea un módulo ventas.py que permita registrar ventas de productos, calcular el total vendido y mostrar un reporte con las ventas realizadas. Importa el módulo en un programa que permita interactuar con un sistema de ventas.

18. Simulador de Ahorros

Escribe un módulo ahorros.py que permita simular el crecimiento de una cuenta de ahorros a lo largo del tiempo, tomando en cuenta una tasa de interés anual. Luego, importa el módulo en un programa que le pida al usuario la cantidad inicial, la tasa de interés y el número de años.

19. Módulo de Manipulación de Archivos

Crea un módulo manipulacion_archivos.py que contenga funciones para leer, escribir y contar palabras en un archivo de texto. Luego, importa este módulo en un programa que le permita al usuario interactuar con archivos desde la consola.

20. Sistema de Reservas

Crea un módulo reservas.py que permita gestionar reservas en un restaurante, registrando nombre, fecha y número de personas. El módulo debe permitir agregar, eliminar y listar reservas. Importa el módulo en un programa que implemente este sistema.

21. Gestión de Inventario

Escribe un módulo inventario.py que contenga funciones para agregar productos al inventario, eliminar productos, y mostrar los productos disponibles con su cantidad y precio. Importa el módulo en un programa principal que permita gestionar un inventario de tienda.

22. Calculadora de Facturas

Crea un módulo facturas.py que permita gestionar facturas con detalles de productos y precios. El módulo debe calcular el total a pagar, incluyendo impuestos. Importa el módulo en un programa que permita generar facturas para los clientes.

23. Sistema de Encuestas

Crea un módulo encuestas.py que permita crear una encuesta con preguntas de opción múltiple y almacenar las respuestas de los usuarios en un archivo de texto. Importa el módulo en un programa principal que permita ejecutar encuestas.

24. Juego de Trivia

Desarrolla un módulo trivia.py que contenga preguntas y respuestas de trivia. El programa principal debe seleccionar aleatoriamente preguntas del módulo y permitir al usuario responderlas, mostrando su puntaje al final.

25. Sistema de Gestión de Vehículos

Crea un módulo vehiculos.py que permita gestionar una lista de vehículos con marca, modelo, año y precio. Importa el módulo en un programa principal que permita agregar, eliminar y mostrar vehículos disponibles en un concesionario.
